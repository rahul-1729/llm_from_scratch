```python
file_path = "./assets/the-verdict.txt"
urllib.request.urlretrieve(url,file_path)
```

```python
import re 
with open("./assets/the-verdict.txt","r") as f:
    text = f.read()
vocab = re.split(r'([,.:;?_!"()\']|--|\s)', text)
vocab = [item.strip() for item in vocab if item.strip()]
vocab
```

```python
len(vocab)
```

```python
results ={token:integer for integer,token in enumerate(vocab)}
results
i=0
for item in enumerate(vocab):
    i=i+1
    print(item)
    if(i>50):
        break
```
